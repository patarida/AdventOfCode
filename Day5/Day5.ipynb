## Part 2:
After the rearrangement procedure completes, what crate ends up on top of each stack?

- In part 1, they can move 1 crate at a time. (FILO)
- In part 2, they can move many crates at a time, so the the crates order remains the same. (FIFO)

```python
# prepare the input

stack1 = ['F','H','B','V','R','Q','D','P']
stack2 = ['L','D','Z','Q','W','V']
stack3 = ['H','L','Z','Q','G','R','P','C']
stack4 = ['R','D','H','F','J','V','B']
stack5 = ['Z','W','L','C']
stack6 = ['J','R','P','N','T','G','V','M']
stack7 = ['J','R','L','V','M','B','S']
stack8 = ['D','P','J']
stack9 = ['D','C','N','W','V']
stacks = [stack1, stack2, stack3, stack4, stack5, stack6, stack7, stack8, stack9]


with open ("input.txt", 'r') as f:
    data = f.readlines()
data = data[10:]
for i in range(len(data)):
    data[i] = data[i].strip()
    
def extractNumber(string):
    split_string = string.split()
    num_crate = split_string[1]
    from_stack = split_string[3]
    to_stack = split_string[5]
    print(int(num_crate), int(from_stack), int(to_stack))
    return int(num_crate), int(from_stack), int(to_stack)
```

```python
# 
for datum in data:
    num_crate, from_stack, to_stack = extractNumber(datum)
    if (len(stacks[from_stack-1]) == 0): continue
            
    # This part will be different from part 1.
    print(stacks[to_stack-1] )
    print(stacks[from_stack-1] )
    print(stacks[from_stack-1][-(num_crate):])
    stacks[to_stack-1] = stacks[to_stack-1] + stacks[from_stack-1][-(num_crate):]
    stacks[from_stack-1] = stacks[from_stack-1][:-(num_crate)]
    print(stacks[from_stack-1] )
    print(stacks[to_stack-1] )
        
print(stacks)
for stack in stacks:
    print(stack[-1], end= "")
```
